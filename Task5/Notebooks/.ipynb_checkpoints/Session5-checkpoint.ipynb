```python
#5.6
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.concatenate((a,b)))
print("The result is a array")

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(np.concatenate((a,b))) #2x4 matrix
print(np.concatenate((a,b), axis = 0)) #2x4 matrix again.
print(np.concatenate((a,b), axis = 1)) #Rotates the the matrix, as in puts the rows together in one row so first row a and first row b together.
```

```python
#5.9
import numpy as np
import imageio
from matplotlib import pyplot as plt

m101 = imageio.imread("../Images/m101BW.jpg")

fig = plt.figure()
frame = fig.add_subplot(1,1,1)
print(m101.shape, m101.dtype)
frame.contour(m101, levels=np.percentile(m101, [90,95,99]), colors=["purple","orchid","pink"])
plt.imshow(m101, cmap = "jet")

m102 = np.where(m101>=90, m101, 0)
fig = plt.figure()
frame = fig.add_subplot(1,1,1)
#print(m102.shape, m102.dtype)
frame.contour(m102)
plt.imshow(m102, cmap ="jet")
```
