```python
#3.10
from math import pi


def cylinder():
    r = float(input("radius in meters please"))
    h = float(input("height in meters please"))
    A = 2*pi*(r**2) + 2*pi*r*h
    V = pi*h*(r**2)
    print(f"The area of the cylinder is: {A} m^2.")
    print(f"The volume of the cylinder is: {V} m^3.")
    return(A,V) #Returning A, V so they can be used outside of the function.

cylinder()

#Calculating the ratio
ratio = A/V
print(f"The ratio Area/Volume is: {ratio}.")
```
