```python
#4.4
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp

#Making the distribution.
b=np.random.normal(loc=185, scale = 8, size = 100000)

#Calculating standard deviation and mean using numpy functions.
bstd=np.std(b)
print(bstd)
bmean = np.mean(b)
print(bmean)

#Checking if the results are similar enough.
if abs(bmean-185)<0.5 and abs(bstd-8)<0.5:
    print("Similar!")
else:
    print("Try again.")

#Plotting
plt.hist(b)
plt.axvline(bmean, marker="o", color="darkblue", markersize=3, label=f"mean={bmean}") #Plots the vertical line for the mean.
plt.axvspan(bmean - bstd, bmean + bstd, color = "orchid", alpha =0.3, label = f"std={bstd}") #Plots the shaded area.
plt.legend()

#Scipy
h = float(input("What's the length?")) #Input for the indicated height.
cdf = sp.stats.norm.cdf(h, loc=bmean, scale=bstd) #First one is upperboundary.
P = (1 -cdf)*100 #Percentile calculation.
print(f"The probability of any values lower or equal to {h} is: {P}.")

up = sp.stats.norm.cdf((h+bstd), loc=bmean, scale=bstd)
low = sp.stats.norm.cdf((h-bstd), loc=bmean, scale=bstd)
P2= (up-low)*100
print(F"The probability of any values witin 1 standard deviation range is: {P2}.")
```
