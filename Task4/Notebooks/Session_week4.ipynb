```python
#4.14

import numpy as np

#Defining the function
def funcfilter(func, seq):
    """Takes the data given for sequence (seq) and makes it a numpy array,
    while the filter takes out everything that is less or equal to 0, afterwards
    it prints the outcome of the function applied to the filtered array."""
    array = np.asarray(seq)
    filt = array[array>0]
    return func(filt)

#Testing with given example 1.
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)

#Testing with given example 2.
s = 100
f = np.log10
result = funcfilter(f,s)
print("Result log10", result)
```

```python
#4.16
import scipy as sp
import numpy as np
from math import sqrt
from astropy.cosmology import Planck18

#Constants
H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0

z = float(input("Please give your value for z")) #Input for upperbound.

#Defining the full function.
def func(x):
    H = H0*sqrt(Om0*(1+x)**3+Ode0)
    t = 1/((1+x)*H)
    return t

conversion = (3.1710e-17 * 3.0856776e19) #First the conversion factor for seconds to Gyr and then times the conversion factor for the s^-1
integral, error = sp.integrate.quad(func,0,z)
result = conversion * integral
res_error = conversion * error
#Ending.
print(f"The result is {result} +- {res_error} Gyr.")
```

```python
#4.21
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm

#Making the normal distributions.
N = 100000
X = np.random.normal(loc = 0, scale = 1, size = N)
Y = np.random.normal(loc = 2, scale = 1, size = N)


#Calculating both the means and the standard deviations.
meanx = np.mean(X)
meany = np.mean(Y)
sdx = np.std(X)
sdy = np.std(Y)


#Calculating the Probability Density Function.
value = np.linspace(-5,7,1000) #Gives the interval and how many points we want in between.
pdfX = norm.pdf(value, meanx, sdx) #Calculates the Probability Density with a build in function.
pdfY = norm.pdf(value, meany, sdy)

#Histogram plots.
plt.hist(X, bins=30, density = True, alpha=0.5, label = "X", color ="Orchid", edgecolor="black") #Array, bins, needed for plotting PDF, transparant, color, edge.
plt.hist(Y, bins=30, density = True, alpha=0.5, label = "Y", color ="purple", edgecolor="black")

#PDF plots.
plt.plot(value, pdfX, label = "PDF x", color = "black", lw = 2, linestyle = ":")#Array, array 2, label, color, linewidth, linestyle.
plt.plot(value, pdfY, label = "PDF y", color = "black", lw = 2, linestyle = "--")

#Rest of the plot.
plt.xticks(np.arange(-6,8,1)) #Making the ticks.
plt.xlabel("Values") #X-axis label
plt.ylabel("Probability Density") #Y-axis label
plt.grid(linestyle = "--", axis = "y") #Plots a grid with -- lines only from the y-axis.
#Making the arrows:
#\mathcal{}
plt.annotate(r"$\mathcal{N}(0,1)$",xy=(-2,0.15),xytext=(-5,0.145), arrowprops = dict(arrowstyle="simple", color="purple", lw=3))
plt.annotate(r"$\mathcal{N}(2,1)$",xy=(4,0.15),xytext=(6, 0.145), arrowprops = dict(arrowstyle="simple", color="blue", lw=3))
#The rotation, placement, arrowstyle, color and then linewidth.

plt.legend()
plt.title("1D Histogram + PDF")
plt.show()



# Compute the 2D histogram
H, xedges, yedges = np.histogram2d(X, Y, bins=30, density=True)
#H is for the pixel value.

# Plot the 2D histogram
plt.figure() #Makes a new figure
plt.pcolormesh(xedges, yedges, H.T, cmap="Blues") # coloured grid for the "pixels"
plt.colorbar(label="Probability Density") # side bar

# Compute bin centers for contours
xcenters = 0.5 * (xedges[:-1] + xedges[1:]) # takes center of bins
ycenters = 0.5 * (yedges[:-1] + yedges[1:])
Xgrid, Ygrid = np.meshgrid(xcenters, ycenters) # makes the grids for x and y

# Use histogram values for contours
Z = H.T # makes sure bins are not rotated 90 degreees by transposing bins

# Contour levels
Z_flat = Z.ravel() # flattens multidimensional array into a single row
Z_sort = np.sort(Z_flat)[::-1] # flips array
Z_cumsum = np.cumsum(Z_sort) # cumulative sum
Z_cumsum /= Z_cumsum[-1]  # divide by last aka normalise

# Define the contour levels corresponding to 1σ, 2σ, and 3σ
levels = [Z_sort[np.searchsorted(Z_cumsum, p)] for p in [0.68, 0.95, 0.997]]#Preparing the contour lines. (searchsorted: "Find indices where elements should be inserted to maintain order")
levels_use = levels[::-1] # flip array around (back to original)

# Plot contours
cs = plt.contour(Xgrid, Ygrid, Z, levels=levels_use,
                 colors="red", linewidths=1.5) # define countour with previous grids and pretty

# Add labels to contours (this will add labels like "1σ", "2σ", "3σ" on the plot)
plt.clabel(cs, inline=True, fontsize=12, fmt={levels_use[0]:"3σ", levels_use[1]:"2σ", levels_use[2]:"1σ"})
#add onto the cs, IN the contour lines, fontsize, overides the original values with new values for in the line.

# Add grid, labels, and title and ticks
#Change tick rate on x axis
plt.xticks(np.arange(-2,7,1))
plt.yticks(np.arange(-2,7,1))

plt.grid(linestyle="--", linewidth=0.7)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2D Histogram + Contour")

# Show the plot
plt.show()
```
