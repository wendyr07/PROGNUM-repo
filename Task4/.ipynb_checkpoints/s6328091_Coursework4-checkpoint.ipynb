### Monte Carlo Error Propagation (4 pt) <font color='red'><b>COURSEWORK</b></font>

One of the classical Physic Lab experiment is to use pendulum to calculate the local gravitational acceleration, ``g``, from the measurements of the pendulum length ``L`` and the oscillation period ``T``.

The formula is expressed as follows:

$$ g=\frac{4\pi^2 L}{T^2} $$

You have got a set of measurements of $L=1.00\pm0.01\ \mathrm{(m)}$ and $T=2.01\pm0.02\ \mathrm{(s)}$.

* Like in Physics Lab, use Tyler expansion described above to propagate the error. Write your derivation in Markdown.

* Calculate the error of the derived ``g``.

* Apply Monte Carlo error propagation to estimate the value and uncertainty of ``g``.

* Make appropriate comments.

* Compare the two results.

$$g = \frac{4\pi^2L}{T^2}$$
$$\Delta g = \sqrt{(\frac{\partial g}{\partial L}\Delta L)^2 + (\frac{\partial g}{\partial T}\Delta T )^2}$$
$$(\frac{\partial g}{\partial L} \Delta L)^2 = (\frac{4\pi^2}{T^2}\Delta L)^2$$
$$(\frac{\partial g}{\partial T} \Delta T)^2 = (\frac{-8\pi^2}{T^3}\Delta T)^2$$
However in the last one we can remove the minus since we square that part, giving us:
$$(\frac{\partial g}{\partial T} \Delta T)^2 = (\frac{8\pi^2}{T^3}\Delta T)^2$$
Comebining those two partial formulas gives us the general error formula:
$$\Delta g = \sqrt{(\frac{4\pi^2}{T^2}\Delta L)^2 + (\frac{-8\pi^2}{T^3}\Delta T)^2}$$
Filling in the given errors above gives the final formula:
$$\Delta g = \sqrt{(\frac{4\pi^2}{T^2}0.01)^2 + (\frac{8\pi^2}{T^3}0.02)^2}$$

````python
#Imports.
from math import pi, sqrt

#Calculating Delta g:
L, L_err = 1.00, 0.01 #In meters.
T, T_err = 2.01, 0.02 #In seconds.

dg = sqrt(((4*pi**2)/(T**2)*L_err)**2 + ((8*pi**2)/(T**3)*T_err)**2) #Calculates the error of g with the formula above.
print(f"The error of g is: {dg:.1f} ms⁻²") #Prints the outcome.
````

The Monte Carlo error propagation formula is:
$$\sigma^{2}_{f} = (\frac{\partial f}{\partial x})^2 \sigma^{2}_{x} + (\frac{\partial f}{\partial y})^2 \sigma^{2}_{y} + ... $$
So in this case:
$$\sigma^{2}_{g} = (\frac{\partial f}{\partial L})^2 \sigma^{2}_{L} + (\frac{\partial f}{\partial T})^2 \sigma^{2}_{T} + ... $$

````python
from math import pi, sqrt #Importing the needed mathematical parts.

def Monte_err(L,sig_L,T,sig_T): #Defining the function.
    a = (4*pi**2)/(T**2) #Partial derivative of g with respect to L.
    b = (8*pi**2*L)/(T**3) #Partial derivative of g with respect to T.
    sig_g = sqrt( a**2 * sig_L**2 + b**2 * sig_T**2) #Gets the error of g by taking the squareroot of the error formula.
    return(sig_g) #Returns the value for the error.


print(f"The error using the Monte Carlo error propagation of g is: {Monte_err(1.00,0.01,2.01,0.02):.1f} ms⁻²") #Prints the error of g with the input of the question.
````

Both the Taylor expansion calculation and the Monte Carlo error propagation formula give the same errors.

### Complex roots of a quadratic equation (4 pt) <font color='red'><b>COURSEWORK</b></font>
  
Use the information in the notebook about complex numbers to write a program that can calculate the (complex) roots of a quadratic equation with the quadratic formula:

$$ y=ax^2+bx+c $$

The requirement of this program is as follows:

* User can supply constants ``a``, ``b`` and ``c`` with Python function `input()`.
* The program can deal with complex roots.
* Your program must print the equation with its constants and its roots.
* The solutions must be inserted to the original equation to prove that $f(x)=0$.

**Answer**:

````python
#Inputting the values.
from math import sqrt

a = float(input("Please fill in the a: "))
b = float(input("Please fill in the b: "))
c = float(input("Please fill in the c: "))
print(f"The formula is: y = {a}x² + {b}x + {c}") #Printing the first formula.

D = b**2 - 4*a*c #Calculates the discriminant.

if D > 0: #D>0 means there are 2 solutions.
    print("There are two solutions:")
    x1 = (-b + sqrt(D))/(2*a) #Calculates the first value of x.
    x2 = (-b - sqrt(D))/(2*a) #Calculates the second value of x.
    print(f"x₁ = {x1:.2f} and x₂ = {x2:.2f}.") #Prints the outcomes and shortens them to two decimals maximum for the print specifically.
elif D == 0: #If D=0 there is exactly 1 solution.
    print("There is one solution:")
    x = -b/(2*a) #Calculates said solution.
    print(f"x = {x:.2f}") #Prints the value of x, again shortened to 2 decimals.
else: #If D<0 there are no real solutions so we have to use complex notation.
    #Solutions are: x = -b/2a +- i sqrt(abs(D))/2*a
    print("There are no real solutions, however the complex solutions are:")
    real = -b/(2*a) #Calculates the real number for the solutions.
    imaginary = sqrt(abs(D))/(2*a) #Calculates the imaginary parts for the soltion.
    x1 = complex(real, imaginary)  #Puts together the real and imaginary parts as a complex number.
    x2 = complex(real, -imaginary) #Puts together the real and imaginary parts as a complex number.
    print(f"x₁ = {x1} and x₂ = {x2}.") #Prints the results.

#Checking the outcomes:
if D == 0:
    print(f"Checking if x = {x} is the solution by substituting it into the formula: y = {a}x² + {b}x + {c}.")
    y = a * x**2 + b * x + c #Calculates the y.
    print(f"y = {y}.")
    if y == 0: #If y = 0 the solution is correct.
        print(f"The solution, x = {x}, is correct since y = 0.")
    if y != 0: #If y is not 0 the solution is not correct.
        print(f"The solution, x = {x}, is incorrect since y is not 0.")
else:
    print(f"Checking if x₁ = {x1} and x₂ = {x2} are the solutions by substituting them into the formula: y = {a}x² + {b}x + {c}.")
    y1 = a * x1**2 + b * x1 + c #Calculating the first y.
    y2 = a * x2**2 + b * x2 + c #Calculating the the second y.
    print(f"y₁ = {y1} and y₂ = {y2}.")
    #y for x1.
    if y1 == 0: #If y1 is 0 the solution is correct.
        print(f"The solution, x₁ = {x1}, is correct since y₁ = 0.")
    else: #y1 is not 0 so something went wrong and x1 is not a solution to the formula.
        print(f"The solution, x₁ = {x1}, is incorrect since y₁ is not 0.")
    #y for x2.
    if y2 == 0: #If y2 is 0 the solution is correct.
        print(f"The solution, x₂ = {x2}, is correct since y₂ = 0.")
    else: #y2 is not 0 so something went wrong and x2 is not a solution to the formula.
        print(f"The solution, x₂ = {x2}, is incorrect since y₂ is not 0.")
````

### Area under a Gaussian (6 pt) <font color='red'><b>COURSEWORK</b></font>

A one dimensional Gaussian function is given by:

$$f(x) = A\ e^{\frac{-{(x-x_0)}^2}{2 \sigma^2}} + z_0$$

$A$ is the amplitude, $x_0$ is the position of the peak, $\sigma$ is the width of the peak
and $z_0$ is the offset in $y$.  

This function can be written in Python as follows:

```python
def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0
```

1. Set in the calling environment:
```python
    A = 1.0
    x0 = 0.0
    sig = 2.0
    z0 = 0.0
```
2. Plot this function between -10 and 10 with 200 samples.
3. Use function `scipy.integrate.quad()` to find the area between ``x=0`` and ``x=2.5``.
4. Use function `matplotlib.pyplot.fill_between()` to mark that area in your plot.
5. Calculate the area between $-\infty$ and $\infty$ ($\approx 5.01$).
6. Given is the integral for the total area under the curve:

    $$\int _{-\infty }^{\infty }Ae^{-(x-x_0)^{2}/(2\sigma^{2})}\,\mathrm{d}x=A \sigma \cdot {\sqrt {2\pi }}$$

   Calculate this area for the values of the parameters given above and prove with ``quad()`` that the two are the same.

7. Make the last part an executable script, called ``gaussarea``. The user can enter values for ``A``, ``x0``, ``sig`` and ``z0`` and the integration limits. The programme then calculates the area under the curve between the integration limits using function ``quad()``. The programme must includes a plot of the curve, the integration area, and the value of that area in a legend.
8. Insert the generated plot in a separate markdown cell below your code cell, and upload your script to GitHub. Provide your GitHub website link in the notebook too for TA to check. 
   
**Answers**:   

````python
import numpy as np
from matplotlib import pyplot as plt
from scipy import integrate
from math import sqrt, pi

#Distribution variables.
A = 1.0
x0 = 0.0
sig = 2.0
z0 = 0.0

#Integration limits:
xlo = 0
xhi = 2.5

#Gauss:
def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0

#Basics:
X = np.linspace(xlo -5,xhi +5,200)
Y = gauss(X, A, x0, sig, z0)
X_between = np.linspace(xlo, xhi, 50)
Y_between = gauss(X_between, A, x0, sig, z0)

inte_gral = integrate.quad(lambda x: gauss(x, A, x0, sig, z0),xlo,xhi)
inte, err = inte_gral

#Plotting the graph:
plt.plot(X,Y, label = "Distribution")
plt.fill_between(X_between, Y_between, color = "orchid", label =f"integrated area is: {inte:.2f}", alpha = 0.5)
plt.title("Area under a Gaussian")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.legend()

xlo, xhi = -np.inf, np.inf
print(f"The area under the graph between -infinity and infinity is: {integrate.quad(lambda x: gauss(x, A, x0, sig, z0), xlo, xhi)}.")

print(f"The outcome of the right part of the equation is: {1.0*2.0*sqrt(2*pi)}.")
````

### Calculate integral of a user supplied function (6 pt) <font color='red'><b>COURSEWORK</b></font>

Write a programme that calculates the integral of any function from the user input using ``eval()``, which is a string representing a function and a variable (floating number) for which the function will be evaluated.

$$ f(x)=x^4+e^{\sin{x}+\cos{x}} $$

Make the function robust with try/except clauses. Demonstrate exceptions for wrong expressions and unknown functions in an expression (NameError) etc. Print messages that tell a user what they have done wrong!

Make the above code an executable script called ``interate_user_function``, and upload to GitHub.

* Create a separate cell, and use SymPy expression to calculate the integral of the formula below for $x=0$ and $x=\pi$.

````python
from numpy import sin, cos, exp, pi
from scipy import integrate
import numpy as np

#example func_str = "x**4+exp(sin(x)+cos(x))" 
func_str = input("Please input a function") #The string function.
xlo = 0 #Lower bound.
xhi = pi #Upper bound.

def f(x): #Defining the function needed for both monte and quad.
    return eval(func_str, {"x": x, "sin": sin, "cos": cos, "exp": exp, "pi": pi})  #Returning the function and the extra clause makes it safer since it doesn't have to search every python enviroment.

#Used from a session quiz.
def monte(f, a, b):
    x = np.random.uniform(a, b, 10000) #Takes points in the range given later on.
    Y = eval(f, {"x": x, "sin": sin, "cos": cos, "exp": exp, "pi": pi})  # evaluate f at all sampled points
    result = (b - a) * np.mean(Y)
    return result

try:
    #Quad integration:
    inte, err = integrate.quad(f, xlo, xhi) #Integrates the function and separates the tuple outcome to two variables.
    print(f"The result of the scipy.integrate.quad is: {inte}.") #Prints quad.
    print(f"The result of the Monte Carlo function is: {monte(func_str, xlo, xhi)}.") #Prints monte.
#First except clause:
except NameError: 
    print("Error: unknown function or variable used, try a different formula.") #Will appear when tan(x) is used or things like that.
#Second except clause:
except SyntaxError:
    print("Error: invalid mathematical expression, check brackets and python format.") #Will appear when the function is not written right, like missing a bracket.
````

### Plotting Hertzsprung–Russell (HR) diagram (6 pt) <font color='red'><b>COURSEWORK</b></font>

``Hertzsprung-Russell (HR) diagram``, which indicates the stellar evolution, is an important astrophysical plot graphing stellar luminosity (brightness, absolute magnitude) against surface temperature (or colour). This diagram clearly states the evolutionary pathways of stars and where they are in their lifetime. With this, we classify several important stages of a star's life -- main sequence stars, giants, or white dwarfs. 

In this exercise, a combined APOGEE/Gaia star catalog has been prepared for you to create an HR diagram -- download the catalogue from the BrightSpace: ``APOGEE+Gaia_stars.fits``

* Firstly, let's select data with good quality and robust measurements using the following criteria:

* The HR diagram contains

    * $y$-axis: absolute magnitude -- $M_G$


**<span style="color:red">Warning</span>**: note that from observation, we only have apparent magnitude (i.e. observed magnitude). To convert into absolute magnitude requires **Distance Modulus**: 

$$ M_G = m_G + 5\log10{(\mathrm{parallax}/1000)} + 5 $$

* Apply a rough binary classification scheme to separate giant branch and main sequence (MS) stars.
    * Giant branch: $ M_G<3.5 $ and $(BP-RP)>0.8$
    * MS: assume the rest of the data is MS

* It seems that we do not have sufficient samples in the White Dwarf (WD) region. Use the following criteria to draw vertical and horizontal lines, as well as define a box indicating the location of WD on the HR diagram.
    * $(BP-RP)\leq0.8$ and $ M_G>10 $ (very faint)
 
* Make a scatter plot and customise it to be look like the plot shown here.



````python
"""
Before this is graded, I want to acknowledge the fact that this plot looks nothing like the desired plot.
If have tried multiple things and I could not get the same look as it. I know the arrows are not where they're
supposed to be, however I still wanted to show I am able to add them and I still tried to get them to show the correct
data sets. I have asked multiple people and the astronomy groupchat all had the same problems with the data set giving 
a lot of points and weird looking outcomes. I just wanted to let you know this in advance.
"""

#Importing stuff.
import numpy as np
from matplotlib import pyplot as plt
from astropy.io import fits

# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')

# Access the primary data (in this case, it is a Table)
data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names

#Getting the good data:
mask = (
    (data['STARFLAG'] == 0) &
    (data['ASPCAPFLAG'] == 0) &
    (data['SNR'] > 10) &
    (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) &
    (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG']))
)
masked_data = data[mask] #Masking the data.

#Taking 80% of the first masked data.
N = len(masked_data) #Defines variable N as the number of data points.
indices = np.random.choice(N, int(0.8 * N), replace=False) #Picks random points for N, with a size of 80% of N, and doesn't replace meaning no double indexes.
sample = masked_data[indices] #Takes for every picked index the values from masked_data.

#Some calculations for the data points:
color = sample['GAIAEDR3_PHOT_BP_MEAN_MAG'] - sample['GAIAEDR3_PHOT_RP_MEAN_MAG'] #Gets the GB - RB for the color (x-axis)
distance = 1000 / sample['GAIAEDR3_PARALLAX'] #Gets the distance needed for the absolute magnitude calculation.
MG = sample['GAIAEDR3_PHOT_G_MEAN_MAG'] - 5*np.log10(distance) + 5 #Calculates the absolute magnitude.

#Masks:
g_mask = (color > 0.8) & (MG<3.5) #Masks the giant branch.
ms_mask = ~g_mask #Masks everything not in the giant branch.

#Plotting the figure:
plt.figure()
plt.title("HR Diagram")
plt.scatter(color[ms_mask], MG[ms_mask], color = "b", label = "Main Sequence", s = 2)
plt.scatter(color[g_mask], MG[g_mask], color = "r", label = "Giant Branch", s=2)
plt.fill_betweenx([10, 20], 0, 0.8, color='lightgray', label = "White Dwarf Region") #Fills in the white dwarf region according to the set bound of the excersise.
plt.xlabel("BP - RP")
plt.xticks(np.arange(0,5,1)) #Making the ticks.
plt.xlim(-0.25,4) #Puts limits for the x-axis to make it look more like the example.
plt.ylabel(r"$M_g$")
plt.ylim(15,-3) #Puts limits for the y-axis to make it look more like the example.
plt.annotate(r"MS",xy=(1,8),xytext = (0.25, 9.25), arrowprops = dict(arrowstyle="simple", color="b", lw=1)) #Arrow 1.
plt.annotate(r"GB",xy=(0.5,0),xytext = (-0.25, 0), arrowprops = dict(arrowstyle="simple", color="r", lw=1)) #Arrow 2.
plt.annotate("WD", xy=(0.25,15)) #Puts the text WD in the white dwarf section.
plt.legend()
plt.show()
````
