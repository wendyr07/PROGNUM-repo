## Task 1 – Star Classification (2pt)

You have measured surface temperatures of several stars. Your goal is to classify each star according to its spectral type. 
* Print the spectral type of each star in the dictionary
* Then, allow the user to input any star's name in the dictionary
* After input a star's name, your code prints its spectral type

**Spectral Types:**

| Type | Temperature (K) |
|------|----------------|
| O    | > 30,000       |
| B    | 10,000 – 30,000|
| A    | 7,500 – 10,000 |
| F    | 6,000 – 7,500  |
| G    | 5,200 – 6,000  |
| K    | 3,700 – 5,200  |
| M    | < 3,700        |

**Star Data (dictionary):**

```python
stars = {
    "Sirius": 9940,
    "Sun": 5800,
    "Proxima Centauri": 3042
}
```

````python
#Star Data (dictionary):

stars = {
    "Sirius": 9940,
    "Sun": 5800,
    "Proxima Centauri": 3042
}

#Makes empty dictionary
spectral = {
}

#Variable input
names = input("Name of the planet please.")
stars[names] = int(input("The temperature of the planet please."))

#Loopje
for star in stars:
    T = stars[star]
    if T >= 30000:
        spectral[star] = "O"
    elif T < 30000 and T >= 10000:
         spectral[star] = "B"
    elif T < 10000 and T >= 7500:
        spectral[star] = "A"
    elif T < 7500 and T >= 6000:
        spectral[star] = "F"
    elif T < 6000 and T >= 5200:
        spectral[star] = 'G'
    elif T < 5200 and T >= 3700:
        spectral[star] = 'K'
    else:
        spectral[star] = 'M'
    print(f"The spectral type of star {star} is {spectral[star]}.")
````

## Task 3 - Habitable Zone Check (4pt)

The habitable zones for a star depends on its luminosity and is given as a range in **AU** (astronomical units). Given known stellar radius and surface temperature, the stellar luminostiy can be calculated with *Stefan–Boltzmann law*: 

$$ L = 4\pi R^2 \sigma T^4,
$$ 

where L, R, and T represent the stellar lumninosity, radius, and surface temperature, respectively, and $\sigma$ is the Stefan–Boltzmann constant. 

The habitable zone is discussed relative to our Sun. In solar unit, $\frac{L}{L_{\odot}}=(\frac{R}{R_{\odot}})^2(\frac{T}{T_{\odot}})^4$, and the habitable distance: $$ D_\mathrm{h} = \sqrt{\frac{L}{L_{\odot}}}\ \mathrm{(AU)} $$


```python
# Unit: solar radius
radii = {
    "Sirius": 1.71,
    "Sun": 1,
    "Proxima Centauri": 0.154,
    "Beta Centauri": 8.4
}
```

The stars listed above, each has one or more planets at different distances to its star as listed below. The habitable zone of a star can be estimated by $$ D_\mathrm{inner} = \sqrt{\frac{D_h^2}{1.06}}$$ $$ D_\mathrm{outer} = \sqrt{\frac{D_h^2}{0.36}}$$


```python
planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}
```

* Calculate the habitable zone for each star.
* Loop through each star and its planets.
* Check whether each planet is within the habitable zone of their host star.
* Using **f-string** to print your result in the following format for each star.

> Star: Sun is G-type star\
> Planet at 0.39 AU: Not Habitable\
> Planet at 0.72 AU: Not Habitable\
> Planet at 1.0 AU: Habitable\
> Planet at 1.80 AU: Not Habitable

````python
from math import sqrt

# Unit: solar radius
radii = {
    "Sirius": 1.71,
    "Sun": 1,
    "Proxima Centauri": 0.154,
    "Beta Centauri": 8.4
}
planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}

#del stars[names]

#L = 4pi R^2 sigma T^4

for star in stars:
    L_solar = (radii[star]**2)*(stars[star])**4
    Dh = sqrt(L_solar)
    Din = sqrt(Dh**2 / 1.06)
    Dout = sqrt(Dh**2 / 0.36)
    print(f"Star: {star} is a {spectral[star]}-type star")
    for distance in planets[star]:
        if Din <= distance <=Dout:
            print(f"Planet at {distance} AU: Habitable")
        else:
            print(f"Planet at {distance} AU: Not Habitable")
    
````
